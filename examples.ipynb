## Cell cycle

```julia
rn = @reaction_network begin
    k1, C + Mp --> C + M
    k2, Cp + M --> C + M
    k3, M + W --> Mp + W
    k4, M + W --> M + Wp
    k5, C --> Cp
    k6, Wp --> W
end
```

```julia
@time sd_result = steady_state_degree(rn)
```

```julia
@time bound_result = lower_bound_of_maximal_positive_steady_state_count(rn)
```

```julia
F = steady_state_system(rn)
```

```julia
grc_result = generic_root_count(F)
```

```julia
bound_result = lower_bound_of_maximal_positive_root_count(F)
```

```julia
# Verify the result for a given choice of b and h
a_spec = bound_result.a_spec
b_spec = bound_result.b_spec
h = bound_result.h
lower_bound_of_maximal_positive_root_count_fixed_a_b_h(F, a_spec, b_spec, h)
```

## The HHK network

```julia
rn = @reaction_network begin
    k1, HK00 --> HKp0
    k2, HKp0 -->  HK0p
    k3, HK0p --> HKpp  
    k4, HK0p  + Hpt --> HK00 + Hptp
    k5, HKpp  + Hpt --> HKp0 + Hptp
    k6, Hptp  --> Hpt
end
```

```julia
@time steady_state_degree(rn)
```

```julia
@time lower_bound_of_maximal_positive_steady_state_count(rn)
```

```julia
# Verify the result for a given choice of b and h
F = steady_state_system(rn)
a = [84, 46, 30, 13, 23, 68]
b = [59, 34]
h = [834, 131, 91, 217, 253, 498]
lower_bound_of_maximal_positive_root_count_fixed_a_b_h(F, a, b, h)
```

## Triangle network

```julia
rn = Catalyst.@reaction_network begin
    k1, 3*X1 + 2*X2 --> 6*X1
    k2, 3*X1 + 2*X2 --> 4*X2 
    k3, 4*X2 --> 3*X1 + 2*X2 
    k4, 6*X1 -->  4*X2
end
```

```julia
@time steady_state_degree(rn)
```

```julia
@time lower_bound_of_maximal_positive_steady_state_count(rn)
```

```julia
F = steady_state_system(rn)
A = matrix(ZZ, [[2, 3]]);
@time toric_root_bound(A, F)
```

```julia
@time toric_lower_bound_of_maximal_positive_root_count(A, F)
```

## 2-site phosphorylation

```julia
rn = @reaction_network begin
    @parameters k1 k2 k3 k4 k5 k6 k7 k8 k9 k10 k11 k12
    @species E(t) F(t)  S0(t) S1(t) ES0(t) FS1(t) S2(t) ES1(t) FS2(t)
  k1, S0 + E --> ES0
  k2, ES0  --> S0 + E
  k3, ES0  --> S1+E
  k4, S1 + F  --> FS1
  k5, FS1  --> S1 + F
  k6, FS1 --> S0 + F
  k7, S1 + E --> ES1
  k8, ES1 --> S1 + E
  k9, ES1 --> S2 + E
  k10, S2 + F  -->FS2
  k11, FS2 --> S2 + F
  k12, FS2 --> S1 + F
end 
```

```julia
@time steady_state_degree(rn)
```

```julia
@time lower_bound_of_maximal_positive_steady_state_count(rn)
```
